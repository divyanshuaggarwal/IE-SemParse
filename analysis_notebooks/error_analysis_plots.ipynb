```python
df = pd.read_csv("/content/drive/MyDrive/IndicSemParse/error_analysis_filtered/domain_wise_f1_results.csv")

df
```

```python
new_df = df.groupby("model").mean().reset_index()
```

```python
ibart = new_df.T.iloc[:, :1]

mbart = new_df.T.iloc[:, 1:2]
```

```python

ibart['model'] = "indicBART-M2O"

mbart['model'] = "mBART-M2O"
```

```python
ibart = ibart.iloc[1:].reset_index()
mbart = mbart.iloc[1:].reset_index()

display(ibart)
display(mbart)
```

```python
ibart.columns = ['domain', 'f1', 'model']
mbart.columns = ['domain', 'f1', 'model']

new_df = pd.concat([ibart, mbart])
```

```python
new_df.sort_values([ "f1", "domain", "model"]).reset_index(drop=True).iloc[list(range(0,20, 2))]['domain']
```

```python
import matplotlib.pyplot as plt

from matplotlib import rcParams


# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 22}

# rcParams.update('font', **font)


# figure size in inches
rcParams['figure.figsize'] = 30,15



sns.barplot(x = 'domain',
            y = 'f1',
            hue="model",
            data = new_df,
            order = new_df.sort_values(["f1", "domain", "model"]).reset_index(drop=True).iloc[list(range(0,20, 2))]['domain']
            )

plt.ylim(bottom=0.3)
plt.ylabel("F1 Score", fontsize=40)
plt.xlabel("Domain Name", fontsize=40)
legend = plt.legend(title='Model', fontsize=50)
plt.setp(legend.get_title(),fontsize=50)

plt.title("Domain Wise average F1 score for all languages", fontsize=50)
plt.xticks(fontsize=30)
plt.yticks(fontsize=40)

plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/IndicSemParse/error_analysis_filtered/frame_rareness_results.csv")

df = df.fillna("0")

df.drop(columns=["hi_orig.json"], inplace=True)


df
```

```python
dataset_name_mapping = {
                        "indic-TOP":"IE-multilingualTOP",
                        "itop":"IE-mTOP",
                        "indic-atis": "IE-multiATIS++"
                        }

df['dataset'] = df['dataset'].map(dataset_name_mapping)
```

```python
new_df = df.groupby(['dataset', 'model', 'frame_rareness']).mean().reset_index()
```

```python
# new_df = new_df[new_df['model'] == "mbart-large-50-many-to-one-mmt"]
```

```python
# new_df.drop(columns=["model"], inplace = True)

new_df
```

```python
new_df = new_df.melt(id_vars=["dataset", "model" ,"frame_rareness"],
        var_name="language",
        value_name="f1")
```

```python
new_df
```

```python
new_df['language'] = new_df['language'].apply(lambda x: x[-5:])
```

```python
new_df[new_df['model'] == "mbart-large-50-many-to-one-mmt"]
```

```python
new_df[new_df['model'] == "mbart-large-50-many-to-one-mmt"].sort_values(['frame_rareness', 'f1']).reset_index(drop=True).iloc[list(range(0,110, 22))]['frame_rareness']
```

```python
import matplotlib.pyplot as plt

from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 16,8


order = ['very_large', 'large', 'medium', 'small', 'very_small']

sns.barplot(x = 'frame_rareness',
            y = 'f1',
            hue="dataset",
            data = new_df[new_df['model'] == "mbart-large-50-many-to-one-mmt"],
            order=order
            )


plt.ylim(bottom=0.3)
legend = plt.legend(title='Dataset', fontsize=20)
plt.setp(legend.get_title(),fontsize=20)
plt.ylabel("F1 Score", fontsize=30)
plt.xlabel("Frame Rareness", fontsize=30)
plt.title("Frame Rareness Wise average F1 score for all languages for mBART-M2O", fontsize=30)
plt.xticks([0, 1, 2, 3, 4], [ "Very High", "High", "Medium", "Low", "Very Low"], fontsize=25)

plt.yticks(fontsize=25)

plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/IndicSemParse/error_analysis_filtered/complexity_results.csv")

df = df.fillna("0")

df.drop(columns=["hi_orig.json"], inplace=True)
```

```python
df['dataset'] = df['dataset'].map(dataset_name_mapping)
```

```python
new_df = df.groupby(['dataset', 'model', 'complexity']).mean().reset_index()
```

```python
new_df = new_df.melt(id_vars=["dataset", "model" ,"complexity"],
        var_name="language",
        value_name="f1")
```

```python
new_df['language'] = new_df['language'].apply(lambda x: x[-5:])
```

```python
import matplotlib.pyplot as plt

from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 20,10



sns.barplot(x = 'complexity',
            y = 'f1',
            hue="dataset",
            data = new_df[new_df['model'] == "mbart-large-50-many-to-one-mmt"])

legend = plt.legend(title='Dataset', fontsize=25)
plt.setp(legend.get_title(),fontsize=30)

plt.ylim(bottom=0.45)
plt.ylabel("F1 Score", fontsize=30)
plt.xlabel("Complexity", fontsize=30)
plt.title("Complexity Wise average F1 score for all languages for mBART-M2O", fontsize=30)
plt.xticks([0,1], ['Complex', "Simple"], fontsize=25)
plt.yticks( fontsize=25)

plt.show()
```
